#### Random Numbers 2 - Random, Shuffle and Choice

```python
#Random float between 0 and 1
uniform_floats = rng.random(50)
print(uniform_floats)
```

```python
#Drawing Randomness
import matplotlib.pyplot as plt
x = rng.random(300)
y = rng.random(300)

plt.scatter(x,y)
plt.show()
```

```python
array = np.array([4, 8, 15, 16, 23, 42])
rng.shuffle(array)
print(f'shuffle array: {array}')
```

```python
three_elements_from_array = rng.choice(array, size=3)
print(f'three elements: {three_elements_from_array}')
```

```python
#We can draw normally distributed random number
heights = rng.normal(175, 15, size=100)
print(heights)
```

```python
y = rng.normal(175, 15, size = 1000)
plt.hist(y, bins=100)
plt.title('normal distribution')
plt.show()
```

#### Basic Statistic

```python
#We have seen the mean (average) of a vector
array = np.array([4, 8, 15, 16, 23, 42])
print(np.mean(array))
```

```python
print(np.std(array))

#Variance 
print(f'squared std: {np.std(array) **2}')
print(f'variance: {np.var(array)}')
```

```python
#Median
print(f'medidan: {np.median(array)}')

#Mean
print(f'mean: {np.mean(array)}')
```

```python
import numpy as np

# Defining the vector called my_vector
my_vector = np.array([1, 4, 5, 10, 13, 9, 10])

# YOURE CODE GOES BELOW HERE!

# Find the mean of the vector
mean = np.mean(my_vector)
print(f'mean: {mean}')
# Find the median of the vector
median = np.median(my_vector)
print(f'median: {median}')
# Find the standard deviation of the vector
standard_deviation = np.std(my_vector)
print(f'std: {standard_deviation}')
```
